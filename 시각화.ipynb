```python
import pandas as pd
import openpyxl
import seaborn as sns
import matplotlib.pyplot as plt
from datetime import datetime
```

```python
from matplotlib import font_manager, rc
rc('font', family='AppleGothic')

plt.rcParams['axes.unicode_minus'] = False
```

```python
# 그래프 그리기
plt.plot([1, 2, 3, 4], [10, 20, 25, 30])
plt.title('한글 제목')
plt.xlabel('X 축 레이블')
plt.ylabel('Y 축 레이블')
plt.show()
```

```python
sub_rain_temp_dow = openpyxl.load_workbook('sub_rain_temp_dow.xlsx')
```

```python
sub_rain_temp_dow = sub_rain_temp_dow.active
sub_rain_temp_dow = pd.DataFrame(sub_rain_temp_dow.values)
```

```python
sub_rain_temp_dow_index = sub_rain_temp_dow
sub_rain_temp_dow_index.columns = sub_rain_temp_dow_index.iloc[0]

# 첫 번째 행을 삭제하여 중복을 제거
sub_rain_temp_dow_index = sub_rain_temp_dow_index[1:]
```

```python
sub_rain_temp_dow_index.reset_index(drop=True, inplace=True)
```

```python
sub_rain_temp_dow_index
```

```python
rainy_days = sub_rain_temp_dow_index[sub_rain_temp_dow_index['강수 여부'] == 1]['sum']
non_rainy_days = sub_rain_temp_dow_index[sub_rain_temp_dow_index['강수 여부'] == 0]['sum']
```

```python
# Seaborn을 사용한 히스토그램 시각화
sns.set(style='whitegrid')
plt.figure(figsize=(10, 5))

sns.histplot(rainy_days, bins=20, color='blue', label='Rainy Days', kde=True)
sns.histplot(non_rainy_days, bins=20, color='green', label='Non-Rainy Days', kde=True)

plt.xlabel('Ridership')
plt.ylabel('Frequency')
plt.title('Ridership Distribution on Rainy vs. Non-Rainy Days')
plt.legend()
plt.show()
```

```python
# Seaborn을 사용한 히스토그램 시각화
sns.set(style='whitegrid')
plt.figure(figsize=(10, 5))

sns.histplot(rainy_days, bins=20, color='blue', label='Rainy Days', kde=True)
sns.histplot(non_rainy_days, bins=20, color='green', label='Non-Rainy Days', kde=True)

plt.xlabel('Ridership')
plt.ylabel('Frequency')
plt.title('Ridership Distribution on Rainy vs. Non-Rainy Days')
plt.legend()

# 각 데이터 집합의 평균에 선 추가
plt.axvline(rainy_days.mean(), color='blue', linestyle='dashed', linewidth=2, label='Mean (Rainy Days)')
plt.axvline(non_rainy_days.mean(), color='green', linestyle='dashed', linewidth=2, label='Mean (Non-Rainy Days)')

plt.legend()
plt.show()
```

```python
# 그래프 설정
plt.figure(figsize=(12, 5))

# 히스토그램 그리기
plt.subplot(1, 2, 1)
sns.histplot(rainy_days, bins=20, color='blue', label='Rainy Days', kde=True)
sns.histplot(non_rainy_days, bins=20, color='green', label='Non-Rainy Days', kde=True)
plt.xlabel('Ridership')
plt.ylabel('Frequency')
plt.title('Ridership Distribution on Rainy vs. Non-Rainy Days')
plt.legend()

# 박스 플롯 그리기
plt.subplot(1, 2, 2)
sns.boxplot(data=[rainy_days, non_rainy_days], palette=['blue', 'green'])
plt.xticks([0, 1], ['Rainy Days', 'Non-Rainy Days'])
plt.xlabel('Days')
plt.ylabel('Ridership')
plt.title('Box Plot of Ridership on Rainy vs. Non-Rainy Days')

plt.tight_layout()
plt.show()
```

```python

# 지하철 역 목록 가져오기
station_list = sub_rain_temp_dow_index['subway_station_name'].unique()

# 그래프 설정
plt.figure(figsize=(12, 6))

# 각 역에 대한 꺾은선 그래프 그리기
for station in station_list:
    station_data = sub_rain_temp_dow_index[sub_rain_temp_dow_index['subway_station_name'] == station]
    
    # 시간대별 승차량 데이터 추출
    time_range = ['05~06', '06~07', '07~08', '08~09', '09~10', '10~11', '11~12', '12~13', '13~14', '14~15', '15~16', '16~17', '17~18', '18~19', '19~20', '20~21', '21~22', '22~23', '23~24']
    ridership = [station_data[time].sum() for time in time_range]
    
    # 꺾은선 그래프 그리기
    plt.plot(time_range, ridership, label=station)

# 그래프 레이아웃 설정
plt.xlabel('Time Range')
plt.ylabel('Ridership')
plt.title('Ridership by Time for Different Subway Stations')
plt.xticks(rotation=45)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')

# 그래프 표시
plt.tight_layout()
plt.show()
```

```python
# 그래프 설정
plt.figure(figsize=(12, 6))

# 요일별 히스토그램 그리기
plt.subplot(1, 2, 1)
sns.histplot(data=sub_rain_temp_dow_index, x='요일', palette='Set2')
plt.xlabel('Day of the Week')
plt.ylabel('Ridership')
plt.title('Ridership Histogram by Day of the Week')

# 요일별 박스 플롯 그리기
plt.subplot(1, 2, 2)
sns.boxplot(data=sub_rain_temp_dow_index, x='요일', y='sum', palette='Set2')
plt.xlabel('Day of the Week')
plt.ylabel('Ridership')
plt.title('Ridership Box Plot by Day of the Week')

plt.tight_layout()
plt.show()
```

```python
# 그래프 설정
plt.figure(figsize=(12, 6))

# 요일별 승차인원 히스토그램 그리기
plt.subplot(1, 2, 1)
sns.histplot(data=sub_rain_temp_dow_index, x='sum', hue='요일', multiple='stack', palette='Set2')
plt.xlabel('Ridership')
plt.ylabel('Frequency')
plt.title('Ridership Histogram by Day of the Week')

# 요일별 박스 플롯 그리기
plt.subplot(1, 2, 2)
sns.boxplot(data=sub_rain_temp_dow_index, x='요일', y='sum', palette='Set2')
plt.xlabel('Day of the Week')
plt.ylabel('Ridership')
plt.title('Ridership Box Plot by Day of the Week')

plt.tight_layout()
plt.show()
```

```python

# 그래프 설정
plt.figure(figsize=(12, 6))

# 요일별 승차인원 히스토그램 그리기
plt.subplot(1, 2, 1)
sns.histplot(data=sub_rain_temp_dow_index[sub_rain_temp_dow_index['요일'] == 'Monday'], x='sum', kde=True, label='Monday', color='blue')
sns.histplot(data=sub_rain_temp_dow_index[sub_rain_temp_dow_index['요일'] == 'Tuesday'], x='sum', kde=True, label='Tuesday', color='green')
sns.histplot(data=sub_rain_temp_dow_index[sub_rain_temp_dow_index['요일'] == 'Wednesday'], x='sum', kde=True, label='Wednesday', color='red')
sns.histplot(data=sub_rain_temp_dow_index[sub_rain_temp_dow_index['요일'] == 'Thursday'], x='sum', kde=True, label='Thursday', color='purple')
sns.histplot(data=sub_rain_temp_dow_index[sub_rain_temp_dow_index['요일'] == 'Friday'], x='sum', kde=True, label='Friday', color='orange')
sns.histplot(data=sub_rain_temp_dow_index[sub_rain_temp_dow_index['요일'] == 'Saturday'], x='sum', kde=True, label='Saturday', color='brown')
sns.histplot(data=sub_rain_temp_dow_index[sub_rain_temp_dow_index['요일'] == 'Sunday'], x='sum', kde=True, label='Sunday', color='pink')

plt.xlabel('Ridership')
plt.ylabel('Density')
plt.title('Ridership Distribution by Day of the Week')
plt.legend()

# 요일별 박스 플롯 그리기
plt.subplot(1, 2, 2)
sns.boxplot(data=sub_rain_temp_dow_index, x='요일', y='sum', palette='Set2')
plt.xlabel('Day of the Week')
plt.ylabel('Ridership')
plt.title('Ridership Box Plot by Day of the Week')

plt.tight_layout()
plt.show()
```

```python

# 그래프 설정
plt.figure(figsize=(10, 6))

# 요일별 승차인원 평균 계산
weekday_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
weekday_avg_ridership = sub_rain_temp_dow_index.groupby('요일')['sum'].mean().reindex(weekday_order)

# 꺾은선 그래프 그리기
plt.plot(weekday_avg_ridership.index, weekday_avg_ridership.values, marker='o', linestyle='-', color='b')
plt.xlabel('Day of the Week')
plt.ylabel('Average Ridership')
plt.title('Average Ridership by Day of the Week (Line Chart)')

plt.xticks(rotation=45)
plt.grid(True)

plt.show()
```
